```python
import os
import sys
import math
import pickle
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle
```

```python
%matplotlib inline
r = np.random.rand(50)
t = np.random.rand(50) * np.pi * 2.0
x = r * np.cos(t)
y = r * np.sin(t)

def get_circle(x, y, r=0.25, color='slateblue'):
    return Circle((x, y), r, facecolor=color, linewidth=3, alpha=1.0)

fig, ax = plt.subplots(figsize=(6, 6))
HAZARD_SIZE = 0.25
rect = Rectangle((-2., -1), HAZARD_SIZE*6, HAZARD_SIZE*6, color='red')
ax.add_patch(rect)
PILLAR_SIZE = 0.25
for x in range(3):
    pos_x = -0.5 + 2*PILLAR_SIZE*x
    for j in range(3):
        pos_y = 0.25 - 2*PILLAR_SIZE*j
        circle = get_circle(pos_x, pos_y)
        ax.add_patch(circle)
for j in range(9):
    pos_y = 1.75 - 2*PILLAR_SIZE*j
    circle = get_circle(-2, pos_y)
    ax.add_patch(circle)
    circle = get_circle(2, pos_y)
    ax.add_patch(circle)
for i in range(7):
    pos_x = -1.5 + 2*PILLAR_SIZE*i
    circle = get_circle(pos_x, 2.)
    ax.add_patch(circle)
    circle = get_circle(pos_x, -2.5)
    ax.add_patch(circle)
ax.set_frame_on(False)
plt.xticks([]), plt.yticks([])
plt.xlim(-2.25, 2.25)
plt.ylim(-2.75, 2.25)
plt.savefig('maze.png', dpi=300, bbox_inches='tight')
```
